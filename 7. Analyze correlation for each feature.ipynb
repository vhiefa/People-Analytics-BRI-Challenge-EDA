```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

I plot correlation map between each feature so that I can find the more important feature and drop unimportant feature. Feature selection may influence the performance of Machine Learning

```python
df_train = pd.read_csv('./6.normalized_train.csv')
df_test = pd.read_csv('./6.normalized_test.csv')
```

```python
f = plt.figure(figsize=(19, 15))
corr = df_train.corr()
corr = corr[corr>.85]
plt.matshow(corr, fignum=f.number)
plt.xticks(range(df_train.shape[1]), df_train.columns, fontsize=14, rotation=90)
plt.yticks(range(df_train.shape[1]), df_train.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

Attribute "person_level" and "job_level" has high correlation value (above 0.9). In dataset, those attributes are a little similar (have same tendency) each other. It makes sense because "person_level" and "job_level" are related. Employee with high "job_level" is usually has high "person_level" as well.  So, there will be no much different between both of it. I will also drop one of these data.



```python
df_train = df_train.drop(columns=['job_level'])
df_test = df_test.drop(columns=['job_level'])
```

```python
f = plt.figure(figsize=(19, 15))
corr = df_train.corr()
corr = corr[corr>.85]
plt.matshow(corr, fignum=f.number)
plt.xticks(range(df_train.shape[1]), df_train.columns, fontsize=14, rotation=90)
plt.yticks(range(df_train.shape[1]), df_train.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

```python
df_train.to_csv("7.selectedfeature_train.csv", index=False)
df_test.to_csv("7.selectedfeature_test.csv", index=False)
```
