```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
from numpy import nan

import matplotlib.pyplot as plt
```

We convert all outlier data into "nan" instead, in the next step all "nan" data will be converted into mean or median. Here we only analyze the outlier data from train dataset. Because we want to analyze based on the class. While the test dataset has no label/class.

```python
df_train = pd.read_csv('./3.num_train.csv')
```

```python
print(df_train.shape)
```

```python
by_class = df_train.groupby('Best Performance')
df_train_positive = by_class.get_group(1)
df_train_negative = by_class.get_group(0)
print(df_train_positive.shape)
print(df_train_negative.shape)
```

```python
def calculateIQRvalue(x) :
    Q3, Q1 = np.nanpercentile(x, [75 ,25])
    IQR = Q3 - Q1
    lower_range = Q1-1.5*IQR
    upper_range = Q3+1.5*IQR
    return(Q1,Q3,IQR,lower_range,upper_range)
```

# job_duration_in_current_job_level

```python
sns.boxplot(x='Best Performance', y='job_duration_in_current_job_level', data=df_train)
_,_,_,lower_pos, upper_pos = calculateIQRvalue(df_train_positive['job_duration_in_current_job_level'])
_,_,_,lower_neg, upper_neg = calculateIQRvalue(df_train_negative['job_duration_in_current_job_level'])
print(lower_pos, upper_pos)
print(lower_neg, upper_neg)
```

```python
df_train.loc[(df_train['job_duration_in_current_job_level'] > upper_pos) & (df_train['Best Performance'] == 1), 'job_duration_in_current_job_level'] = nan
df_train.loc[(df_train['job_duration_in_current_job_level'] < lower_pos) & (df_train['Best Performance'] == 1), 'job_duration_in_current_job_level'] = nan
df_train.loc[(df_train['job_duration_in_current_job_level'] > upper_neg) & (df_train['Best Performance'] == 0), 'job_duration_in_current_job_level'] = nan
df_train.loc[(df_train['job_duration_in_current_job_level'] < lower_neg) & (df_train['Best Performance'] == 0), 'job_duration_in_current_job_level'] = nan
```

```python
sns.boxplot(x='Best Performance', y='job_duration_in_current_job_level', data=df_train)
```
