```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('./7.selectedfeature_train.csv')
df_test = pd.read_csv('./7.selectedfeature_test.csv')
```

```python
def plot_distribution(df_train, name, categorical=False):
    if not categorical:
        sns.displot(df_train, x=name, kind="kde",
              hue='Best Performance')
        plt.title(f'{name} -> Best Performance', pad=15)
        plt.show()
    else:
        sns.displot(df_train, x=name,
              hue='Best Performance')
        plt.title(f'Train: {name} -> Best Performance', pad=15)
        plt.show()
```
