```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
from numpy import nan

import matplotlib.pyplot as plt
```

We convert all invalid data into "nan" instead, in the next step all "nan" data will be converted into mean or median

In this process, I analyze the IQR value based on the merge data between train and test dataset. This is because the training data is more general than testing data itself. the testing data has no label, so just in case it's not general enough.

```python
df_train = pd.read_csv('./4.train_without_outlier.csv')
df_test = pd.read_csv('./3.num_test.csv')
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_merge_data = pd.concat([df_train,df_test])
```

```python
def calculateIQRvalue(x) :
    Q3, Q1 = np.nanpercentile(x, [75 ,25])
    IQR = Q3 - Q1
    lower_range = Q1-1.5*IQR
    upper_range = Q3+1.5*IQR
    return(Q1,Q3,IQR,lower_range,upper_range)
```

# job_duration_in_current_job_level

```python
sns.boxplot(x=df_merge_data['job_duration_in_current_job_level'])
_,_,_,lower_rang, upper_rang = calculateIQRvalue(df_merge_data['job_duration_in_current_job_level'])
print(lower_rang, upper_rang)
```

```python
df_test.loc[df_test['job_duration_in_current_job_level'] > upper_rang, 'job_duration_in_current_job_level'] = nan
df_test.loc[df_test['job_duration_in_current_job_level'] < lower_rang, 'job_duration_in_current_job_level'] = nan
sns.boxplot(x=df_test['job_duration_in_current_job_level'])
```
